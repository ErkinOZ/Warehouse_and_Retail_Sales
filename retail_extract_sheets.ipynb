```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('Warehouse_and_Retail_Sales.csv')
```

```python
df
```

```python
sales_by_item_type = df.groupby("ITEM TYPE").agg({
    "RETAIL SALES": "sum",
    "RETAIL TRANSFERS": "sum",
    "WAREHOUSE SALES": "sum"
}).reset_index()
```

```python
sales_by_item_type
```

```python
df["Year-Month"] = df["YEAR"].astype(str) + "-" + df["MONTH"].astype(str).str.zfill(2)

sales_by_month = df.groupby("Year-Month").agg({
    "RETAIL SALES": "sum",
    "WAREHOUSE SALES": "sum"
}).reset_index()
```

```python
sales_by_month
```

```python
top_suppliers = df.groupby("SUPPLIER")["RETAIL SALES"].sum().reset_index()
top_suppliers = top_suppliers.sort_values(by="RETAIL SALES", ascending=False).head(5)
```

```python
top_suppliers
```

```python
sales_comparison = df.groupby("Year-Month").agg({
    "RETAIL SALES": "sum",
    "RETAIL TRANSFERS": "sum",
    "WAREHOUSE SALES": "sum"
}).reset_index()
```

```python
sales_comparison
```

```python
top_items = df.groupby(["ITEM CODE", "ITEM DESCRIPTION"])["RETAIL SALES"].sum().reset_index()
top_items = top_items.sort_values(by="RETAIL SALES", ascending=False).head(10)
```

```python
top_items
```

```python
sales_by_supplier = df.groupby("SUPPLIER").agg({
    "RETAIL SALES": "sum",
    "WAREHOUSE SALES": "sum"
}).reset_index()
```

```python
sales_by_supplier
```

```python
supply_frequency = df.groupby(["Year-Month", "SUPPLIER"]).agg({
    "RETAIL TRANSFERS": "sum",
    "WAREHOUSE SALES": "sum"
}).reset_index()
```

```python
supply_frequency
```

```python
yearly_sales = df.groupby("YEAR").agg({
    "RETAIL SALES": "sum",
    "WAREHOUSE SALES": "sum"
}).reset_index()
```

```python
yearly_sales
```

```python
with pd.ExcelWriter("sales_analysis.xlsx") as writer:
    sales_by_item_type.to_excel(writer, sheet_name="Sales by Item Type", index=False)
    sales_by_month.to_excel(writer, sheet_name="Sales by Month", index=False)
    top_suppliers.to_excel(writer, sheet_name="Top Suppliers", index=False)
    sales_comparison.to_excel(writer, sheet_name="Sales Comparison", index=False)
    top_items.to_excel(writer, sheet_name="Top Items", index=False)
    sales_by_supplier.to_excel(writer, sheet_name="Sales by Supplier", index=False)
    supply_frequency.to_excel(writer, sheet_name="Supply Frequency", index=False)
    yearly_sales.to_excel(writer, sheet_name="Yearly Sales", index=False)
```
